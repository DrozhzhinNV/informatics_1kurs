–§–æ—Ä–º–∞—Ç –≤–≤–æ–¥–∞:
* –≤ –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ —É–∫–∞–∑–∞–Ω–æ N - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–∞–º–Ω–µ–π –≤ –æ–∑–µ—Ä–µ
* –¥–∞–ª–µ–µ –∏–¥—É—Ç N —Å—Ç—Ä–æ–∫ —Å –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º–∏ –∫–∞–º–Ω–µ–π. –ü–µ—Ä–≤—ã–π –∫–∞–º–µ–Ω—å –≤ —Å–ø–∏—Å–∫–µ - –∫–∞–º–µ–Ω—å –§–µ–¥–∏, –≤—Ç–æ—Ä–æ–π - –∫–∞–º–µ–Ω—å –§–∞–Ω–∏.

–§–æ—Ä–º–∞—Ç –≤—ã–≤–æ–¥–∞:
* –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ —Å —Ç–æ—á–Ω–æ—Å—Ç—å—é –¥–æ 3 —Ü–∏—Ñ—Ä –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π, —Ä–∞–≤–Ω–æ–µ –º–∏–Ω–∏–º–∞–∫—Å–Ω–æ–º—É —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—é –º–µ–∂–¥—É –∫–∞–º–Ω—è–º–∏ –§–µ–¥–∏ –∏ –§–∞–Ω–∏.
–ü—Ä–∏–º–µ—Ä
–≤–≤–æ–¥:

18 5

–≤—ã–≤–æ–¥:
1.414

```python
import sys

count = int(input())
mas = []
for i in range(count):
    mas.append(list(map(int, input().split())))
min = sys.maxsize

for i in mas:
    for j in mas:
        dist = ((i[0] - j[0])**2 + (i[1] - j[1])**2)**0.5
        if dist < min and dist != 0:
            min = dist

print(round(min, 3))
```
